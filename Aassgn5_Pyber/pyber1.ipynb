```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

 # Import our data into pandas from CSV
 


 # Combine the data into a single dataset
pyberone = pd.merge(ride_data, city_data, how="left", on=["city", "city"])

# # Display the data table for preview
pyberone.head()                        
```

```python
urban_cities = pyberone.loc[pyberone["type"] =="Urban"]

urban_countt = urban_cities.groupby("city")["ride_id"].count()

urban_avg = urban_cities.groupby("city")["fare"].mean()
urban_avg

urban_drtotal = urban_cities.groupby("city")["driver_count"].sum()
urban_drtotal

urban_drtot = urban_cities.groupby("city")["driver_count"].mean()
urban_drtot

```

```python
suburban_cities = pyberone.loc[pyberone["type"] =="Suburban"]
suburban_cities
suburban_countt = suburban_cities.groupby("city")["ride_id"].count()

suburban_avg = suburban_cities.groupby("city")["fare"].mean()
suburban_avg

suburban_drtotal = suburban_cities.groupby("city")["driver_count"].sum()
suburban_drtotal
suburban_drtot = suburban_cities.groupby("city")["driver_count"].mean()
suburban_drtot
```

```python
rural_cities = pyberone.loc[pyberone["type"] =="Rural"]
rural_cities
rural_countt = rural_cities.groupby("city")["ride_id"].count()
rural_countt
rural_avg = rural_cities.groupby("city")["fare"].mean()
rural_avg
#rural_drtotal = rural_cities.groupby("city")["driver_count"].mean()
#rural_drtotal
rural_drtotal1 = rural_cities.groupby("city")["driver_count"].sum()
rural_drtotal1
```

# Bubble plot of ride sharing data 

```python
#plt.subtitle('Pyber Ride Sharing Data')
plt.scatter(urban_countt, urban_avg, s=urban_drtotal*.4, linewidths=1, c="lightcoral", edgecolor ="black",label="urban", alpha=0.70)
plt.scatter(suburban_countt,suburban_avg, s=suburban_drtotal*.4, linewidths=1, marker="o", c="skyblue", edgecolor ="black",label="suburban", alpha=0.80)
plt.scatter(rural_countt, rural_avg, s=rural_drtotal1,linewidths=1, marker="o",c="gold",edgecolor="black",label="rural", alpha=0.80)
plt.title("Pyber Ride sharing Data (2016)")
plt.xlabel("Total Number of Rides Per City")
plt.ylabel("Average Fare($)")

lgnd = plt.legend(frameon = True, edgecolor ='black')
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
plt.text(45,35,"Notes:\nCircle size correlates with driver count per City")
plt.grid()
plt.show()
```

# Percent of total Fare by city type

```python
#Total fare by City type

percent_city_fare = pyberone.groupby(["type"])["fare"].sum()
labels = ["Rural","Suburban","Urban" ]
colors = ["gold","skyblue","lightcoral"]
explode = (0, 0, 0.1)
plt.title("% of Total Fares By City Types")
plt.pie(percent_city_fare, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%",shadow=True, startangle=160)
plt.axis("equal")
plt.show()
```

# Percent of total Rides by city type

```python
#Total rides by City type
percent_rides_city = pyberone.groupby(["type"])["ride_id"].count()
labels = ["Rural","Suburban","Urban" ]
colors = ["gold","skyblue","lightcoral"]
explode = (0, 0, 0.1)
plt.title("% of Total Rides By City Types")
plt.pie(percent_rides_city, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%",shadow=True, startangle=140)
plt.axis("equal")
plt.show()
```

# Percent of total Drivers by city type

```python
#% of Total Drivers by City type
percent_drivers_city = city_data.groupby(["type"])["driver_count"].sum()
labels = ["Rural","Suburban","Urban"]
colors = ["gold","skyblue","lightcoral"]
explode = (0, 0, 0.1)
plt.title("% of Total Drivers by City Types")
plt.pie(percent_drivers_city, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%",shadow=True, startangle=140)
plt.axis("equal")
plt.show()
```

# Trends and Analysis 



1. The Bubble Plot indicates that there are more drivers and rides in urban cities. In Urban areas, trips are more likely to be shorter but happen at a higher frequency.  
Where as in suburban and rural areas this is not the case. 
Suburban and Rural areas have more opportunities to expand their business and also grow by hiring more drivers to operate. 

2. Since the Rural cities have few drivers this leads to  fewer rides that in turn, on average, a higher price per ride. 
Since we do not have more information regarding the gas price and the trip distance it is difficult to predict but there is a
chance that Suburban cities can succeed with proper planning and coming up with cost effective ways for hiring and managing 
the fare. 

3. Suburban cities fall in between the two extremes of Rural and Urban. Suburban have a lower ride count than Urban while
having a higher average price per ride.
